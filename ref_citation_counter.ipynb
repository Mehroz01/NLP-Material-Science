```python
def freq_count(file_read):

    data = pd.read_csv(file_read)
    l1=data['entity']
    #return dict(counter)
    #counter=collections.Counter(l1)
    return len(set(list(l1)))
    # 

def citations_count(file_read):

    data = pd.read_csv(file_read)
    l1=data['citations'][0]
    return l1

def images_count(file_read):

    data = pd.read_csv(file_read)
    l1=data['no_of_images'][0]
    return l1
    
#counter=collections.Counter(l1)
# dict(counter)




def get_freq(file_read):
    frequencies={}
    frequencies['composition']=freq_count('/home/mehroz/Mat-Sci/freq_data/compositions/'+file_read[:-4]+'.csv')
    frequencies['processes']=freq_count('/home/mehroz/Mat-Sci/freq_data/processes/'+file_read[:-4]+'.csv')
    frequencies['properties']=freq_count('/home/mehroz/Mat-Sci/freq_data/properties/'+file_read[:-4]+'.csv')
    frequencies['devices']=freq_count('/home/mehroz/Mat-Sci/freq_data/method_devices/'+file_read[:-4]+'.csv')
    frequencies['citations']=citations_count('/home/mehroz/Mat-Sci/freq_data/citations/'+file_read[:-4]+'.csv')
    frequencies['images']=images_count('/home/mehroz/Mat-Sci/freq_data/images/'+file_read[:-4]+'.csv')
 
    return frequencies




#file_read='/home/mehroz/Mat-Sci/processes/processes_data/1-s2.0-S0264127516313090-main.pdf'
#get_freq('')
```
